```python
import pandas as pd
import numpy as np
```

```python
compras_original = pd.read_csv('dataset_group.csv',
sep= ',',
encoding= 'utf-8',
names = ['data','ticket','produto'])                               
```

```python
compras_original.head()
```

```python
compras = pd.crosstab(compras_original.ticket, compras_original.produto, dropna= False)
```

```python
compras
```

```python
#Transformar toda a tabela em 1 ou 0
```

```python
compras[compras>1]=1
compras
```

```python
from matplotlib import cm
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
frequent_itemsets = apriori(compras, min_support=0.3, use_colnames=True)
rules = association_rules(frequent_itemsets, metric ='lift', min_threshold=1)
```
